# using Matplotlib

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=['SAT','SUN','MON','TUE','WES','THU','FRI']
y=[40,30,33,24,55,36,27]
```

```python
plt.plot(x,y,'g',linewidth=5)
plt.show()
```

```python
plt.scatter(x,y)
```

# Matplotlib Histograms.

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

```python
plt.hist(x, weights=y, bins=len(x), color = "y")
plt.xlabel('day')
plt.ylabel('temp')
plt.title('Histograms are used to represent the temperature')
plt.show()
```

# Matplotlib Pie Charts.
With Pyplot, you can use the pie() function to draw pie charts:

```python
m=[20,40,30,10]
mylabels = ["20%", "40%", "30%", "10%"]
plt.pie(m,labels = mylabels,shadow = True)
plt.legend()

plt.show()
```

# Matplotlib Bars


```python
plt.bar(x,y,width = 0.1,color='0')
plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('Bar chart')
plt.show
```

# color Reference : 
'0' as black

'1' as white

'0.8' as light gray

'b' as blue

'g' as green

'r' as red

'c' as cyan

'm' as magenta

'y' as yellow

'k' as black

'w' as white
