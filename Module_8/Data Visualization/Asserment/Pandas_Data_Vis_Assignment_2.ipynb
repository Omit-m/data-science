Let's go through a quick exercise to review the pandas built-in visualization capabilities. Run the first cell to import required libraries and to create a dataset for use in the exercises below. <br>
*`np.random.seed(1)` to get the repeatable results.*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
df = pd.DataFrame(np.random.rand(1000,4),columns=['a', 'b', 'c','d'])
#df = pd.DataFrame(np.random.randn(1000,4),columns=['a', 'b', 'c','d'])
%matplotlib inline
```

**How many entries we have in our dataframe?**

```python
df.info()
```

**Check the head of the dataframe?**

```python
 df.head()
```

**You have created a dataset `df`. Plot `c vs d` as a scatter plot to recreate the plot below. Try to match the color and size of the data points. You should be able to match the size of the figure as well!**

```python
plt.figure(figsize=(7, 3)) 
plt.scatter(df['c'], df['d'], color='g', s=20)
```

** Create a histogram of the 'd' column.**

```python
plt.figure(figsize=(4, 3)) 

plt.hist(df['d'])
plt.ylabel('Frequency')
```

** Change style sheets to set for 'bmh' and add more bins (e.g. 50) to the above plot.**

```python
plt.figure(figsize=(5, 3))
plt.style.use('bmh')
plt.hist(df['d'], bins=50)
plt.ylabel('Frequency')
```

**Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?**<br>

```python
plt.figure(figsize=(5, 3)) 
plt.scatter(df['a'], df['b'], s=df['d']*50, color='green')
```

**Recreate the plot below**

```python
df.plot.hexbin(x='a', y='b', gridsize=30, cmap='Blues', figsize=(4, 3))
```

**Create a boxplot comparing all columns in the data.**<br>
Hint: use `color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray')` and `sym = 'r+'` 

```python
plt.figure(figsize=(4, 3)) 

boxplot = df.boxplot(color=dict(boxes='Green', whiskers='Orange', medians='Blue', caps='Gray'), sym='r+')
```
