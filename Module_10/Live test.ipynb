# Following are some model performance information

Correct answer = 10

Total questions = 30

True Positive = 8

False Positive = 6

False Negative = 4


# Based on the above information, find out the following result.

Accuracy =

Precision =

Recall =

F1 =0


# Answer

Accuracy :

Accuracy=(True Positives / Total questions)
True Positives = 8
Total questions=30
Accuracy =
 ( 8 / 30 )=0.2667

Precision :

Precision = (True Positives / (True Positives + False Positives))
True Positives = 8
False Positives = 6
Precision =
  (8/(8+6))=0.5714


Recall :

Recall=(True Positives / (True Positives + False Negatives))
True Positives = 8
False Negatives = 4
Recall = 
(8/(8+4))=0.6667

F1 Score :

F1 Score= 2*((Precision * Recall ) / ( Precision + Recall ))
Precision = 0.5714
Recall = 0.6667
F1 Score =
 2 * ((0.5714 * 0.6667 ) / (0.5714 * 0.6667 )) = 2 * ( 0.3809 / 1.2381 ) = 0.6154
