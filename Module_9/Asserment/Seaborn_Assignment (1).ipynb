```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Do the following tasks:** 
* set style to '`white`'
* load titanic dataset into '`titanic`', tips into '`tips`' and iris into '`iris`'

```python
sns.set_style('white')
```

```python
titanic = sns.load_dataset('titanic')
tips= sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
plt.figure(figsize=(5,3))
sns.boxplot(x="day", y="tip",
            hue="sex", 
            data=tips)
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python

plt.figure(figsize=(2, 3)) 
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex',height=4)
plt.show()
```

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.** 



```python
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(5, 3))
sns.distplot(titanic['fare'],kde=False,bins=20,color='g')
```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
plt.figure(figsize=(6, 4)) 
sns.boxplot(x="class", y="age", data=titanic,
           hue = 'sex')
plt.legend(loc=1)
```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
sns.jointplot(x='fare', y='age', data=titanic,  height=5)
```

**Recreate the plot below. Use titanic dataset. ** 

```python
plt.figure(figsize=(5, 4)) 

sns.boxplot(x="class", y="age", data=titanic,palette='rainbow')

plt.show()
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
count_data = titanic.groupby(['class', 'sex']).size().reset_index(name='count')


sns.barplot(x="sex", y="count", hue="class", data=count_data)

plt.legend(loc=1)
```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots** 

```python

selected_columns =  titanic[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']]
plt.figure(figsize=(5, 4))
T1=selected_columns.corr()
plt.title('titanic.corr()')

sns.heatmap(T1,annot=True,cmap='coolwarm')
```
