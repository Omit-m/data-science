```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


data_dic = {
    'A': [12, 15, 45, 71],
    'B': [18, 26, 50, 80],
    'C': [32, 48, 78, 93],
    'D': [29, 35, 63, 81]
}

T1=pd.DataFrame(data_dic)


plt.figure(figsize=(5,4))
sns.heatmap(T1,annot=True)
```
